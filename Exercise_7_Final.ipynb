### Section 1

* arrived_trains=10
* departed_trains=10
* include_nonstopping=False

```python
import requests
url1 = "https://rata.digitraffic.fi/api/v1/live-trains?station=HKI&arrived_trains=10&departed_trains=10&include_nonstopping=false"
resp = requests.get(url1)
if resp.status_code == 200:
    train_data = resp.json()
    print(train_data)
```

2 Read the data into the Pandas dataframe without editing, so that the column headings are as follows:

 `cancelled`,
 `commuterLineID`,
 `departureDate`,
 `operatorShortCode`,
 `operatorUICCode`,
 `runningCurrently`,
 `timeTableRows`,
 `timetableAcceptanceDate`,
 `timetableType`,
 `trainCategory`,
 `trainNumber`,
 `trainType`,
 `version`

```python
import requests
import pandas as pd

url1 = "https://rata.digitraffic.fi/api/v1/live-trains?station=HKI&arrived_trains=10&departed_trains=10&include_nonstopping=false"

response = requests.get(url1)
data = response.json()

# Define the required column names
columns = [
    "cancelled",
    "commuterLineID",
    "departureDate",
    "operatorShortCode",
    "operatorUICCode",
    "runningCurrently",
    "timeTableRows",
    "timetableAcceptanceDate",
    "timetableType",
    "trainCategory",
    "trainNumber",
    "trainType",
    "version",
]

df = pd.DataFrame(data, columns=columns)
print(df.head())
```

3 Delete all other columns except `timeTableRows` and `trainNumber`.

```python
df = df[["timeTableRows", "trainNumber"]]
print(df.head())
```

4 After this, extract the values ​​of the `timeTableRows` column from each train separately (the column value for each train contains several JSON objects) and add `trainNumber`, which identifies the train data.

```python
import pandas as pd

df_exploded = df.explode("timeTableRows")

df_timetable = pd.json_normalize(df_exploded["timeTableRows"])

df_timetable["trainNumber"] = df_exploded["trainNumber"].values

print(df_timetable.head())
```

5 Then drop the following columns from the data: `causes`, `differenceInMinutes` and `commercialStop`.

```python
df_timetable = df_timetable.drop(columns=["causes", "differenceInMinutes", "commercialStop"], errors="ignore")

print(df_timetable.head())
```

6 Fill in the missing values ​​of the `actualTime` column in each row with the values of the `scheduledTime` column.

```python
df_timetable["actualTime"] = df_timetable["actualTime"].fillna(df_timetable["scheduledTime"])

print(df_timetable.head())
```

### Section 2

7 Extract the cell values ​​of the `table` element of the `data.html` file using the regex syntax and connect them to the dataframe created in step 1. **Note**: only regex should be used in the task (E.g. the use of Pandas' read_html() method is prohibited)!

```python
import re
import pandas as pd

with open("data.html", "r", encoding="utf-8") as file:
    html_content = file.read()

cell_pattern = re.findall(r"<td>\\s*(.*?)\\s*</td>", html_content)

columns = ["actualTime", "cancelled", "commercialTrack", "countryCode", "scheduledTime", "stationShortCode", "stationUICCode", "trainStopping", "type", "trainNumber"]

rows = [cell_pattern[i:i + len(columns)] for i in range(0, len(cell_pattern), len(columns))]

df = pd.DataFrame(rows, columns=columns)

print(df)
```

### Section 3

8 Make a SQLite database and insert the dataframe obtained as a result of step 2 to the database.

```python
import sqlite3

conn = sqlite3.connect("train_data.db")
cursor = conn.cursor()

cursor.execute(f"""
CREATE TABLE IF NOT EXISTS train_schedule (
    actualTime TEXT,
    cancelled TEXT,
    commercialTrack TEXT,
    countryCode TEXT,
    scheduledTime TEXT,
    stationShortCode TEXT,
    stationUICCode TEXT,
    trainStopping TEXT,
    type TEXT,
    trainNumber TEXT
)""")

df.to_sql("train_schedule", conn, if_exists="replace", index=False)

conn.commit()
conn.close()

print(df)
```
